```python
#Data cleaning and transformations using Pandas
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(10)
```

```python
df['Ship Mode'].unique()
```

```python
#handling null values
df = pd.read_csv('orders.csv',na_values = ['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
#renaming column names to all lower case and underscores instead of spaces

#df.rename(columns = {'Order Id' : 'order_id', 'Ship Mode' : 'ship_mode'}.....becomes tough for more columns)

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns
```

```python
#deriving new columns like discount, sale_price and profit

df['discount'] = df['list_price'] * df['discount_percent'] * 0.01

df['sale_price'] = df['list_price'] - df['discount']

df['profit'] = df['sale_price'] - df['cost_price']

df
```

```python
#dropping unwanted columns

df.drop(columns = ['cost_price','list_price','discount_percent'], inplace = True)

'''inplace = True drops columns from the dataframe, if we dont use it the columns will be dropped only when this 
cell is run that is temporarily'''
```

```python
#dealing with datatypes

df.dtypes
```

```python
#lets convert order_date from object to datetime datatype

df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
df.dtypes
```
